```python
from confluent_kafka import Producer
import json
import time
import random
from datetime import datetime

producer = Producer({'bootstrap.servers': 'kafka:9092'})
topic = "iot_sensors"
print(f" Connected to Kafka broker at kafka:9092 â€” sending data to topic '{topic}'")

def delivery_report(err, msg):
    if err is not None:
        print(f" Delivery failed: {err}")
    else:
        print(f" Delivered to {msg.topic()} [{msg.partition()}] offset={msg.offset()}")

while True:
    data = {
        "device_id": f"sensor_{random.randint(1, 5)}",
        "temperature": round(random.uniform(20.0, 35.0), 2),
        "humidity": random.randint(30, 80),
        "pressure": random.randint(980, 1050),
        "air_quality": random.randint(0, 100),
        "noise_level": random.randint(20, 90),
        "battery_level": random.randint(10, 100),
        "signal_strength": random.randint(-90, -30),
        "latitude": round(random.uniform(29.9, 30.1), 4),
        "longitude": round(random.uniform(31.1, 31.3), 4),
        "timestamp": datetime.utcnow().isoformat()
    }

    producer.produce(
        topic=topic,
        key=data["device_id"],
        value=json.dumps(data),
        callback=delivery_report
    )
    producer.poll(0)

    print(f" Sent: {data}\n")
    time.sleep(3)  
```
